```python
# kne_summary_plot.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import rubin_sim.maf as maf

# Load metric subsets and summary values
metric_subsets = maf.get_metric_subsets('/Users/andradenebula/Documents/Research/Transient_Metrics/Multi_Transient_Metrics_Hub/KNe/KNe_Lynn/metric_subsets.json')
summaries = maf.get_metric_summaries(summary_source='/Users/andradenebula/Documents/Research/Transient_Metrics/Multi_Transient_Metrics_Hub/KNe/KNe_Lynn/summary.h5')


# Select only the kilonova-related metrics from TVS
kne_subset = metric_subsets.loc['TVS KNe']
kne_metrics = kne_subset['metric']

# List of cadences you care about
cadences = ['baseline_v4.3.1_10yrs', 'four_roll_v4.3.1_10yrs']

# Pull the summary values for those cadences and metrics
kne_summaries = summaries.loc[cadences, kne_metrics]
print("\nKilonova Metric Summary Values:")
print(kne_summaries)

# Plot each metric as a bar plot comparing cadences
for metric_name in kne_metrics:
    values = kne_summaries[metric_name]
    for metric_name in kne_metrics:
        values = kne_summaries[metric_name]
        print(f"\nMetric: {metric_name}")
        print(values)
    
        plt.figure()
        values.plot(kind='bar')
        plt.title(metric_name)
        plt.ylabel('Metric Value')
        plt.xlabel('Cadence')
        plt.grid(True, axis='y')
        plt.tight_layout()
        plt.show()


```
